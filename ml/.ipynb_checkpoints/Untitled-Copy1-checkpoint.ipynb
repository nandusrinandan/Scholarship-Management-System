```python
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
plt.style.use('bmh')
```

```python
data=pd.read_csv('DATASET.csv')
```

```python
data
```

```python
data.shape
```

```python
plt.figure(figsize=(16,8))
plt.title('Schoalrship Acceptance')
plt.xlabel('CGPA')
plt.ylabel('Probability of Acceptance')
plt.plot(data['CGPA'])
plt.show()
```

```python
data=data[['CGPA']]
data.head(10)
```

```python
future_days = 25
data['Prediction']=data[['CGPA']].shift(-future_days)
data.head(10)
```

```python
x=np.array(data.drop(['Prediction'],1))[:-future_days]
print(x)
```

```python
#Create the target data set (y) and convert it to a numpy array and get all of the target value except the last 'x' rows/days
y=np.array(data['Prediction'])[:-future_days]
print(y)
```

```python
#Split the data into 75% training and 25% testing
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25)
```

```python
#Create the models
#Create the DECISION TREE REGRESSOR MODEL
tree = DecisionTreeRegressor().fit(x_train,y_train)

#Create the LINEAR REGRESSION MODEL
lr= LinearRegression().fit(x_train,y_train)
```

```python
#Get the last 'x' rows of the feature data set
x_future = data.drop(['Prediction'],1)[:-future_days]
x_future = x_future.tail(future_days)
x_future=np.array(x_future)
x_future
```

```python
tree_prediction = tree.predict(x_future)
print(tree_prediction)
print()
#Show the model linear regression
lr_prediction = lr.predict(x_future)
print(lr_prediction)
```

```python
predictions=tree_prediction

valid = data[x.shape[0]:]
valid['Predictions']=predictions
plt.figure(figsize=(16,8))
plt.title('Scholarship Acceptance Graph')
plt.xlabel('CGPA')
plt.ylabel('Probability of Acceptance')
plt.plot(data['CGPA']) 
plt.plot(valid[['CGPA','Predictions']])
plt.legend(['Orig','Val','Pred'])
plt.show()
```
